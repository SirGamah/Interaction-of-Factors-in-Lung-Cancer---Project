About Dataset

üìå Overview

This dataset has been carefully synthesized to support research in lung cancer survival prediction, enabling the development of models that estimate:

1. Whether a patient is likely to survive at least one year post-diagnosis (Binary Classification).
2. The probability of survival based on clinical and lifestyle factors (Regression Analysis).

The dataset is designed for machine learning and deep learning applications in medical AI, oncology research, and predictive healthcare.

üìú Dataset Generation Process

The dataset was generated using a combination of real-world epidemiological insights, medical literature, and statistical modeling. The feature distributions and relationships have been carefully modeled to reflect real-world clinical scenarios, ensuring biomedical validity.

üìñ Medical References & Sources

The dataset structure is based on well-established lung cancer risk factors and survival indicators documented in leading medical research and clinical guidelines:

  * World Health Organization (WHO) Reports on lung cancer epidemiology.
  * National Cancer Institute (NCI) & American Cancer Society (ACS) guidelines on lung cancer risk factors and treatment outcomes.
  * The IASLC Lung Cancer Staging Project (8th Edition): Standard reference for lung cancer staging.
  * Harrison‚Äôs Principles of Internal Medicine (20th Edition): Provides an in-depth review of lung cancer diagnosis and treatment.
  * Lung Cancer: Principles and Practice (2022, Oxford University Press): Clinical insights into lung cancer detection, treatment, and survival factors.

üî¨ Features of the Dataset
Each record in the dataset represents an individual‚Äôs clinical condition, lifestyle risk factors, and survival outcome. The dataset includes the following features:

1Ô∏è‚É£ Patient Demographics

* Age ‚Üí A key risk factor for lung cancer progression and survival.
* Gender ‚Üí Male and female lung cancer survival rates can differ.
* Residence ‚Üí Urban vs. Rural (impact of environmental factors).

2Ô∏è‚É£ Risk Factors & Lifestyle Indicators

These factors have been linked to lung cancer risk in epidemiological studies:

* Smoking Status ‚Üí (Current Smoker, Former Smoker, Never Smoked).
* Air Pollution Exposure ‚Üí (Low, Moderate, High).
* Biomass Fuel Use ‚Üí (Yes/No) ‚Äì Associated with household air pollution.
* Factory Exposure ‚Üí (Yes/No) ‚Äì Industrial exposure increases lung cancer risk.
* Family History ‚Üí (Yes/No) ‚Äì Genetic predisposition to lung cancer.
* Diet Habit ‚Üí (Vegetarian, Non-Vegetarian, Mixed) ‚Äì Nutritional impact on cancer progression.

3Ô∏è‚É£ Symptoms (Primary Predictors)

These are key clinical indicators associated with lung cancer detection and severity:

* Hemoptysis (Coughing Blood)
* Chest Pain
* Fatigue & Weakness
* Chronic Cough
* Unexplained Weight Loss

4Ô∏è‚É£ Tumor Characteristics & Clinical Features

* Tumor Size (mm) ‚Üí The size of the detected tumor.
* Histology Type ‚Üí (Adenocarcinoma, Squamous Cell Carcinoma, Small Cell Carcinoma).
* Cancer Stage ‚Üí (Stage I to Stage IV).

5Ô∏è‚É£ Treatment & Healthcare Facility

* Treatment Received ‚Üí (Surgery, Chemotherapy, Radiation, Targeted Therapy).
* Hospital Type ‚Üí (Private, Government, Medical College).

6Ô∏è‚É£ Target Variables (Predicted Outcomes)

* Survival (Binary) ‚Üí 1 (Yes) if the patient survives at least 1 year, 0 (No) otherwise.
* Survival Probability (%) (Can be derived) ‚Üí Estimated probability of survival within one year.

### 1. Age Analysis

```python
# Plot histogram of `Age`
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

This plot highlights the importance of age as a factor in lung cancer.

* The impact of genetic vulnerabilities can become more apparent or significant with age due to accumulated environmental exposures and biological changes. Analyzing age alongside genetic data could help identify age-specific genetic risk factors.


```python
# Box plot of `Age`
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Age'])
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

The box plot provides a summary of the age distribution, showing the median, quartiles, and potential outliers.
* Genetic predispositions to lung cancer can manifest at different ages. Some genetic factors might increase risk earlier in life, while others might become more influential with age due to cumulative factors.
* The box plot reinforces the idea that lung cancer affects individuals across a broad age spectrum.

```python
# Plot `Age` vs `Survival`
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survival_1_Year', y='Age', data=df)
plt.title('Age vs Survival')
plt.xlabel('Survival')
plt.ylabel('Age')
plt.show()
```

This box plot compares the age distribution of patients who survived for at least one year (`'Yes'`) versus those who did not (`'No'`). From the plot, it appears that the median age and the overall distribution of ages are quite similar between the two survival groups. The boxes and whiskers for both `'Yes'` and `'No'` survival outcomes largely overlap.
* This plot suggests that age by itself might not be a strong predictor of one-year survival in this dataset. This highlights the need to consider other factors, including genetic vulnerabilities, to better understand survival outcomes.
* While the marginal effect of age on survival appears limited here, it's possible that genetic factors interact with age in complex ways that are not evident in this simple box plot.
* To achieve precision in disease prevention and treatment, we need to move beyond analyzing single factors in isolation. Integrating genetic information with clinical data like age, tumor characteristics, and treatment received is essential for building more accurate predictive models for survival.

```python
# Plot histogram of `Age` colored by `Survival_1_Year`
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Survival_1_Year', kde=True)
plt.title('Distribution of Age by Survival')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram of `Age` colored by `Survival_1_Year`
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Age', hue='Survival_1_Year', kde=True)
plt.title('Distribution of Age by Survival')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.boxplot(x='Survival_1_Year', y='Age', data=df)
plt.title('Age vs Survival')
plt.xlabel('Survival')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```

This stacked histogram provides a more detailed look at the age distribution for both survival groups ('Yes' and 'No').
```python
# One-hot encode all categorical features
X = pd.get_dummies(X, columns=['Gender', 'Smoking_Status', 'Residence',
                               'Air_Pollution_Exposure', 'Biomass_Fuel_Use',
                               'Factory_Exposure', 'Family_History', 'Diet_Habit',
                               'Symptoms', 'Histology_Type', 'Stage', 'Treatment',
                               'Hospital_Type'])
```

**Conclusion on Model Performance for Precision Prevention:**

For the purpose of **Interaction of Factors in Lung Cancer**, a model with high recall is particularly valuable. Identifying individuals who are likely to survive ('Yes') is important for confirming the effectiveness of interventions. However, a high number of false positives (predicting survival when it doesn't occur) can lead to a false sense of security.

The **Random Forest** and **Neural Network** models demonstrate the highest ability to identify actual survivors (high recall). However, their high number of false positives is a concern. The **Gradient Boosting** model offers a better balance between identifying survivors and correctly identifying non-survivors (better trade-off between recall and precision/true negatives). The **Decision Tree** performs the worst overall.

The choice of the "best" model depends on the specific goals of the precision prevention program. If the priority is to identify as many potential survivors as possible for positive reinforcement or to assess the upper bound of intervention effectiveness (and the cost of false positives is less critical), then Random Forest or Neural Network might be preferred due to their high recall. If a more balanced approach is needed, where correctly identifying non-survivors is also important (e.g., for targeting intensive interventions), the Gradient Boosting model might be more suitable.

Regardless of the model, the predicted survival probabilities can be used to stratify risk. Individuals with lower predicted probabilities (even if classified as 'Yes' by models with high false positives) could be flagged for personalized intervention strategies to potentially improve their outcomes, aligning with the principles of precision prevention. Future work could involve fine-tuning these models and incorporating explicit genetic data (if available) to build even more precise predictive models.
